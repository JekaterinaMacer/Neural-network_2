##Библиотека gTTS

Данная библиотека находится в свободном доступе, использование бесплатно.
Документация по ссылке: https://gtts.readthedocs.io/en/latest/

```
import gtts # Импорт библиотеки
from gtts import gTTS # Основной интерфейс для перевода текста в аудио
from IPython.display import Audio # Используем для проигрывания аудио
```

```
# Словарь со всеми доступными языками
# К сожалению, для русского языка доступен только один голос
gtts.lang.tts_langs()
```

```
text = """Ночь, улица, фонарь, аптека,
          Бессмысленный и тусклый свет.
          Живи ещё хоть четверть века —
          Всё будет так. Исхода нет.

          Умрёшь — начнёшь опять сначала
          И повторится всё, как встарь:
          Ночь, ледяная рябь канала,
          Аптека, улица, фонарь."""
tts = gTTS(text, lang='ru') # Преобразовываем текст в аудио
tts.save('blok.mp3') # Сохраняем в формате мр3
```

```
Audio('blok.mp3')
```

###Препроцессинг

*Разбор данного раздела:* https://youtu.be/W-7Nwq47xv0?t=210

Данная библиотека позволяет изменять параметры обработки текста.


Препроцессинг модифицирует входной текст  под конкретные нужды

```
tts = gTTS('РФ - страна с огромной территорией', lang='ru') # Текст без препроцессинга
tts.save('rf.mp3')
Audio('rf.mp3')
```

```
from gtts.tokenizer import pre_processors # Импорт препроцессоров
gtts.tokenizer.symbols.SUB_PAIRS.append(('РФ', 'Российская федерация')) # Добавляем замену 
text = 'РФ - страна с огромной территорией' # Текст
tts = gTTS(text, lang='ru') # Синтезируем речь
tts.save('rf.mp3')
Audio('rf.mp3')
```

Можно сделать даже больше (но лучше не надо)

```
# Добавляем в список замен дополнительные элементы (строка, ее замена)
gtts.tokenizer.symbols.SUB_PAIRS.append(('стр.', 'cтрана'))                        
gtts.tokenizer.symbols.SUB_PAIRS.append(('огр.', 'огромной'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('терр.', 'территория'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('дейст.', 'действительно'))
gtts.tokenizer.symbols.SUB_PAIRS.append(('бол.', 'большая'))
```

```
text = 'РФ - стр. с огр. терр. Ее терр. дейст. бол.' # Плохочиаемый текст, но благодаря заменам, аудио будет звучать нормально
tts = gTTS(text, lang='ru') # Синтезируем речь
tts.save('1.mp3')
Audio('1.mp3')
```

Как правило, препроцессинг используется для замены часто встречающихся аббривиатур и сокращений(стр., РФ, ред. и прочее), а также для задания стиля прочтения пунктуации (Полное игнорирование, игнорирование запятых и т.д.)

###Пример использования

*Разбор данного раздела:* https://youtu.be/W-7Nwq47xv0?t=502

```
# Инициализация клиента
client = texttospeech.TextToSpeechClient()
# Получаем список всех русских голосов
voices = client.list_voices(language_code='ru-RU')
```

```
# language_codes - код языка(Для русского ru-RU)
# name - название голоса
# ssml_gender - пол
# natural_samlple_rate_hertz - Частота дискретизации голоса в Гц
print(voices)
```

```
text = """Ночь, улица, фонарь, аптека,
          Бессмысленный и тусклый свет.
          Живи ещё хоть четверть века —
          Всё будет так. Исхода нет.

          Умрёшь — начнёшь опять сначала
          И повторится всё, как встарь:
          Ночь, ледяная рябь канала,
          Аптека, улица, фонарь."""
          
text_input = texttospeech_v1.types.SynthesisInput(text=text) # Входной текст
```

```
# Выбор голоса
voice = texttospeech_v1.types.VoiceSelectionParams(
    language_code='ru-Ru',
    name = 'ru-RU-Standard-B')
```

```
# Конфигурация аудио
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech_v1.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.8) # Скорость произношения
```

```
response = client.synthesize_speech(input=text_input, voice=voice, audio_config=audio_config) # Синтезируем речь
```

```
# Ответ сохраняем mp3 формате
with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
    print('Файл сохранен под именем output.mp3')
```

```
Audio('output.mp3')
```

### Отличие Wavenet от базовых голосов

###Конфигурация аудио

*Разбор данного раздела:* https://youtu.be/W-7Nwq47xv0?t=876

Данная библиотека позволяет настраивать конфигурацию аудио

```
# Будем использовать женский Wavenet голос
voice = texttospeech_v1.types.VoiceSelectionParams( 
    language_code='ru-Ru',
    name = 'ru-RU-Wavenet-E')
```

```
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9, # Скорость произношения
    pitch = 0, # Высота голоса
    sample_rate_hertz = 24000, # Частота дискретизации
    volume_gain_db = 0) # Усиление громкости в децибелах
```

Разная высота голоса

```
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9, # Скорость произношения
    pitch = -5) # Высота голоса
response = client.synthesize_speech(input=text_input, voice=voice, audio_config=audio_config) # Синтезируем речь
with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
Audio('output.mp3')
```

```
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.9, # Скорость произношения
    pitch = 5) # Высота голоса
response = client.synthesize_speech(input=text_input, voice=voice, audio_config=audio_config) # Синтезируем речь
with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
Audio('output.mp3')
```

```
audio_config = texttospeech_v1.types.AudioConfig(
    audio_encoding=texttospeech.AudioEncoding.MP3, # Формат мр3
    speaking_rate = 0.5) # Скорость произношения
response = client.synthesize_speech(input=text_input, voice=voice, audio_config=audio_config) # Синтезируем речь
with open('output.mp3', 'wb') as out:
    out.write(response.audio_content)
Audio('output.mp3')
```
